```python
import zipfile
import os

# Get all zip files in current directory
for filename in os.listdir('.'):
    if filename.lower().endswith('.zip'):
        folder_name = filename[:-4]  # remove .zip extension
        print(f"Extracting {filename} -> {folder_name}/")

        # Create folder if needed
        os.makedirs(folder_name, exist_ok=True)

        # Extract zip contents
        with zipfile.ZipFile(filename, 'r') as zip_ref:
            zip_ref.extractall(folder_name)

print("Done!")
```

First we bring in the textbook as a pdf and convert all the pages into images for the ocr reader

```python

def pdf_to_images(pdf_path, numberOfPages:int=-1, dpi:int=200):
    doc = fitz.open(pdf_path)
    firstBookImages = {}
    for page_num, page in tqdm(enumerate(doc)):
        if page_num == numberOfPages:
            break
        pix = page.get_pixmap(dpi=200) 
        # type pixmap , dpi 200 is ideal , any less than 100 is very low quality and frankly from 100 to 200 is not much difference but its betteer to be safe than sorry 
        firstBookImages[page_num] = pix
    return firstBookImages
def printImage(image: fitz.Pixmap):
    img = Image.frombytes("RGB", [image.width, image.height], image.samples)
    display(img)
```
